### Resampling 

La méthode SMOTE (Synthetic Minority Over-sampling Technique) est un méthode de rééchantillonnage utilisée pour traiter les problèmes d'asymétrie de données en générant de nouveaux échantillons pour les classes sous-représentées.

SMOTE utilise une approche de génération de données synthétiques pour augmenter la taille de l'ensemble de données de la classe minoritaire. Il fonctionne en sélectionnant un échantillon de la classe minoritaire au hasard et en trouvant ses k plus proches voisins. Pour générer un nouvel échantillon synthétique, il choisit aléatoirement un des k plus proches voisins et calcule un point interpolé entre cet échantillon et l'échantillon sélectionné. Ce point interpolé est utilisé comme un nouvel échantillon synthétique pour la classe minoritaire.

En utilisant cette méthode, SMOTE génère de nouveaux échantillons synthétiques qui sont similaires aux échantillons existants de la classe minoritaire, ce qui permet de conserver les caractéristiques de cette classe tout en augmentant sa taille. Cela permet de réduire l'impact de l'asymétrie de données sur les modèles d'apprentissage automatique et d'améliorer les performances de ces modèles sur les classes minoritaires

```python
def resampling(features, target, resample):
    print('Original dataset shape %s' % Counter(target))
    X, y = resample.fit_resample(features, target)
    print('Resampled dataset shape %s' % Counter(y))
    return X, y
sm = SMOTE(random_state=42, sampling_strategy=0.2)
```

## Choix d'un modèle 

Pour entraîner et comparer différents modèles de machine learning sur notre jeu de données, nous avons utilisé trois algorithmes différents : le 'Dummy Classifier', le 'Random Forest' et 'LGBM'. Nous avons entraîné ces modèles avec les mêmes hyperparamètres et les mêmes données d'entraînement et de validation croisée pour garantir une comparaison juste et équitable. Après l'entraînement, nous avons évalué les performances de chaque modèle en utilisant bank-score. Le meilleur modèle est celui qui a le meilleur score.

```python
my_models2 = { 'Dummy Classifier': DummyClassifier(),
                  'Random Forest': RandomForestClassifier(n_estimators=100),
                'LGBM': LGBMClassifier()}
```

1) Dummy Classifier :\
Le Dummy Classifier est un modèle de classification très simple qui sert principalement de référence pour comparer les performances d'autres modèles plus avancés. Ce modèle fonctionne en attribuant des prédictions aléatoires à chaque observation de l'ensemble de données d'entraînement. Par exemple, dans le cas d'une classification binaire, le Dummy Classifier peut attribuer des prédictions aléatoires de 0 ou de 1 à chaque observation, sans se baser sur les caractéristiques de l'observation. Bien que ce modèle ne soit pas utile pour des prévisions précises, il permet de comparer les performances d'autres modèles en termes de précision, de rappel, de F1-score et d'autres métriques d'évaluation.\


2) Random Forest :\
Le Random Forest est un modèle de classification ensembliste qui combine plusieurs arbres de décision pour améliorer les performances de prédiction. Le modèle fonctionne en créant un grand nombre d'arbres de décision, chacun étant entraîné sur un sous-ensemble aléatoire des données d'entraînement et sur un sous-ensemble aléatoire des caractéristiques. Lors de la prédiction, le modèle effectue une majorité de vote sur les prédictions de chaque arbre pour déterminer la classe finale. Le Random Forest est très efficace pour des ensembles de données de grande taille, avec des caractéristiques complexes et non linéaires.


3) LGBM :\
Le LGBM (Light Gradient Boosting Machine) est un modèle de gradient boosting de type arbre, similaire au Random Forest. Cependant, il utilise une approche de construction d'arbre plus efficace et plus rapide, qui se concentre sur les échantillons de données les plus importants pour améliorer les performances de prédiction. Le modèle utilise également une méthode d'optimisation des gradients pour ajuster les poids de chaque arbre et minimiser la fonction de perte. Le LGBM est très efficace pour les ensembles de données de grande taille et de haute dimensionnalité, avec des caractéristiques non linéaires et interactives.

Nous allons afficher pour chaque modéle la matrice de confusion et le score ROC( Receiver Operating Characteristic)

```python
from prettytable import PrettyTable

def create_confusion_matrix_table(TN, FP, FN, TP):
    table = PrettyTable()
    table.field_names = ["", "Valeur réelle : 0", "Valeur réelle : 1"]
    table.add_row(["Prédiction : 0", TN, FN])
    table.add_row(["Prédiction : 1", FP, TP])
    return table
TN = 'Vrais négatifs (True Negatives, TN)'
FP = 'Faux positifs (False Positives, FP)'
FN = 'Faux négatifs (False Negatives, FN)'
TP = 'Vrais positifs (True Positives, TP)'

create_confusion_matrix_table(TN, FP, FN, TP)
#print(table)
```

#### métrique bancaire

La métrique bankaire consiste à calculer le gain obtenu pour l'ensemble des individus du jeu de données.

Pour cela nous avons fixé un poids arbitraire pour chacune des prédictions relativement à leurs valeurs réelles.

Les valeurs des poids sont les suivantes :  fn_value = 10 - tp_value = 0 - tn_value = 1 - fp_value = 1

De ce fait, les prêts accordés aux individus qui ne sont finalement pas solvables sont dotés d'une pénalisation négative de -10, alors que les prêts accordés aux individus finalement solvables rapportent 1. Ce rapport 10 est totalement arbitraire et il est tout à fait possible de changer ces valeurs à la convenance de l'optique métier. Il faudra cependant relancer l'optimisation des hyper-paramètres du modèle.

```python
def bank_score(y_true, y_pred, fn_value=-10, fp_value=0, tp_value=0, tn_value=1):

    # Matrice de Confusion
    mat_conf = confusion_matrix(y_true,y_pred)
    
    tn = mat_conf[0, 0]
    fn = mat_conf[1, 0]
    fp = mat_conf[0, 1]
    tp = mat_conf[1, 1]
    
    
    Gain_tot = tp*tp_value + tn*tn_value + fp*fp_value + fn*fn_value
    G_max = (fp + tn)*tn_value + (fn + tp)*tp_value
    G_min = (fp + tn)*fp_value + (fn + tp)*fn_value
    
    G_normalized = (Gain_tot - G_min)/(G_max - G_min)
    
    return G_normalized  
```

```python
def model_optim(model, param_grid, X, y, X_valid, y_valid):
    import timeit
    start_time = timeit.timeit()
    with tqdm(total=len(param_grid)) as pbar:
        def grid_search_callback(*args, **kwargs):
            pbar.update()
# votre code ic
    cv = 3
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid,scoring=bank_score, cv=cv)
    grid_search.fit(X, y, callbacks=[grid_search_callback])
    best_param = grid_search.best_params_
    best_model = grid_search.best_estimator_
    best_score = grid_search.best_score_
    model.set_params(**best_param)
    y_pred_grd = best_model.predict_proba(X_valid)[:, 1]    
    end_time = timeit.timeit()
    fpr, tpr, thresholds = roc_curve(y_valid, y_pred_grd)
    print('les meilleur paramètres pour bank_score sont:', best_param)
# Calcul de l'AUC
    roc_auc = auc(fpr, tpr)

# Tracé de la courbe ROC
    plt.plot(fpr, tpr, label='AUC = %0.2f' % roc_auc)
    plt.legend(loc='lower right')
    plt.plot([0,1], [0,1], 'r--')
    plt.xlim([0,1])
    plt.ylim([0,1])
    plt.ylabel('TPR (Sensibilité)')
    plt.xlabel('FPR (Spécificité)')
    plt.show()
    print("Temps d'exécution:", end_time - start_time)
    return best_model, best_param, best_score, y_pred_grd
#best_model_B_S, best_param_B_S, best_score_B_S, y_pred_grd_B_S = model_optim(LGBMClassifier(), param_grid, bank_score,  X_train_sm_test, y_train_sm, X_valid_sm, y_valid_sm)
best_model_B_S, best_param_B_S, best_score_B_S, y_pred_grd_B_S =  model_optim(LGBMClassifier(), param_grid, X_train_test, y_train, X_valid, y_valid)
#y_pred = best_model_B_S.predict(X_valid_sample_sm)
```

```python
def optimize_threshold_bank_score(model, X, y, fn_value=-10, fp_value=1, tp_value=0, tn_value=0):
    y_pred = model.predict_proba(X)[:,1]
    thresholds = np.arange(0, 1.05, 0.005)
    scores = []
    for threshold in thresholds:
        y_pred_bin = (y_pred > threshold).astype(int)
        score = bank_score(y, y_pred_bin, fn_value, fp_value, tp_value, tn_value)
        scores.append(score)
    best_threshold = thresholds[np.argmax(scores)]
    plt.plot(thresholds, scores)
    plt.xlabel('Thresholds')
    plt.ylabel('bank_score')
    plt.show()
    return best_threshold
```

## Précision, rappel et roc_auc


1) Accuracy (précision) :\
L'accuracy (ou précision) est une métrique qui mesure la proportion de prédictions correctes parmi toutes les prédictions effectuées par le modèle. Elle est calculée en divisant le nombre total de prédictions correctes par le nombre total de prédictions effectuées. 


2) Recall (rappel) :\
Le recall (ou rappel) est une métrique qui mesure la proportion de vrais positifs détectés par le modèle parmi tous les vrais positifs dans les données d'entraînement. Elle est calculée en divisant le nombre de vrais positifs détectés par le nombre total de vrais positifs. Le recall est une mesure utile pour les situations où l'on souhaite détecter toutes les observations positives, même si cela entraîne une augmentation des faux positifs.


3) ROC AUC :\
Le ROC AUC (Area Under the Receiver Operating Characteristic Curve) est une métrique qui évalue la capacité du modèle à distinguer les observations positives des observations négatives en fonction de différents seuils de classification. Elle est calculée en traçant la courbe ROC (Receiver Operating Characteristic) qui représente le taux de vrais positifs en fonction du taux de faux positifs pour différents seuils de classification. La zone sous la courbe ROC (ROC AUC) est une mesure de la qualité globale du modèle, où une valeur de 1 correspond à une classification parfaite et une valeur de 0,5 correspond à une classification aléatoire.

```python
def calculate_multi_metrics(clf, x_valid, y_valid, threshold, *metrics):
   
    # liste des scores
    scores = []
    scores_treshold = []
    
    # calcule de la probabilité retourné par le modèle
    valid= clf.predict(x_valid)
    valid_proba = clf.predict_proba(x_valid)[:, 1]
    # calcule de la prédiction en fonction du threshold
    valid_predictions = (valid_proba > threshold)
    valid_predictions = np.array(valid_predictions > 0) * 1
    
    # Calcule le score pour chaque métrique
    for metric in metrics :
        scores.append(metric(y_valid, valid_predictions))
        scores_treshold.append(metric(y_valid, valid))  
    return scores, scores_treshold # retourne la liste des différents scores
```

## Selection de variable 

Shapley Additive exPlanations (SHAP) est une méthode d'interprétation de modèles de machine learning qui fournit des explications de la prédiction en calculant l'importance de chaque caractéristique (feature) dans la prédiction du modèle. Cette méthode est basée sur la théorie des jeux coopératifs, en particulier sur le concept de valeur de Shapley. La valeur de Shapley est une mesure de contribution d'un joueur à un jeu coopératif, qui prend en compte les contributions des autres joueurs dans différentes combinaisons de joueurs.

Dans le contexte de l'interprétation de modèles de machine learning, les caractéristiques (features) du modèle sont considérées comme des joueurs dans un jeu coopératif, où le but est de prédire une valeur de sortie à partir des entrées. Les valeurs de Shapley sont utilisées pour mesurer l'importance de chaque feature dans la prédiction finale du modèle. Les valeurs de Shapley sont calculées en évaluant la différence de prédiction entre l'inclusion et l'exclusion d'une caractéristique donnée, pour toutes les combinaisons possibles de caractéristiques.

La méthode SHAP permet d'obtenir des explications individuelles pour chaque prédiction, en calculant les valeurs de Shapley pour chaque caractéristique de l'entrée individuelle. Ces explications permettent de comprendre comment le modèle a pris sa décision pour chaque entrée, en identifiant les caractéristiques qui ont le plus contribué à la prédiction. SHAP est une méthode très utile pour comprendre les modèles de machine learning complexes, en particulier pour les applications où l'explicabilité est importante, comme la santé, la finance ou le droit.

```python
def feature_importance(model, X):
    feature_importances = model.feature_importances_

    feature_names = X.columns

    # Create a dictionary of feature names and importances
    feature_importances_dict = dict(zip(feature_names, feature_importances))
    sum_of_importance = best_model_B_S.feature_importances_.sum()
    # Sort the feature importances by importance
    sorted_feature_importances = sorted(feature_importances_dict.items(), key=lambda x: x[1], reverse=True)
    list_features = []
    importances = []
    # Print the feature importances
    for feature, importance in sorted_feature_importances:
    #print(f'{feature}: {importance/sum_of_importance:.4f}')
        importances.append(importance/sum_of_importance)
        list_features.append(feature)    

    df_importance_feature  = pd.DataFrame(importances, list_features)
    df_importance_feature = df_importance_feature.reset_index()
    df_importance_feature.columns = ['feature', 'imprtance']
    df_importance_feature = df_importance_feature.sort_values(by='imprtance',ascending=False)


    plt.figure(figsize = (8, 4))
    ax = plt.subplot()
    ax.invert_yaxis()
    ax.grid(visible = True, axis = 'x', linestyle = '-.')
    ax.barh(df_importance_feature['feature'].head(15),
        df_importance_feature['imprtance'].head(15), 
        align = 'center')
    
    ax.set_xlabel('Normalized Importance', fontsize = 12)
    ax.set_title('Feature Importances', fontsize = 14)
    #plt.show()
    return plt.show()
feature_importance(best_model_B_S, X_train)
```
