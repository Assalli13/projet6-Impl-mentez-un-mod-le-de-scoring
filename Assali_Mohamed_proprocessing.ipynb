```python
import numpy as np
def remove_redundant_variables(df, threshold):
    corr_matrix = df.corr()
    col_i = []
    col_j = []
    correlation = []
    n = corr_matrix.shape[0]
    to_remove = []
    for i in range(n):
        for j in range(i):
            corr = abs(corr_matrix.iloc[i, j])
            if corr > threshold:
                col_i.append(corr_matrix.columns[i])
                col_j.append(corr_matrix.columns[j])
                correlation.append(corr)
                to_remove.append(corr_matrix.columns[j])
    to_remove = np.unique(to_remove)
    data_corr = pd.DataFrame([col_i, col_j,correlation]).T
    data_corr.columns = ['col_i', 'col_j', 'corr']
    display(data_corr)
    return to_remove
to_remove = remove_redundant_variables(application_train_A, 0.7)
```

#### Encodage de la varaible 'WEEKDAY_APPR_PROCESS_START'

La fonction polar_projection permet de projeter les jours de la semaine sur un cercle, en utilisant la méthode de codage polar. Cette méthode consiste à transformer chaque catégorie en une paire de coordonnées $(x,y)$ correspondant aux coordonnées polaires de la position de la catégorie sur un cercle. Cette projection peut aider à visualiser les relations entre les catégories. Dans ce code, la fonction prend en entrée un dataframe app et une colonne col contenant les jours de la semaine. Elle retourne le dataframe avec la colonne col remplacée par deux nouvelles colonnes correspondant aux coordonnées polaires du jour de la semaine. Les données de la colonne originale sont ainsi transformées pour pouvoir être utilisées dans les modèles de machine learning.

```python
def polar_projection(app, col):
    # Create map from categories to polar projection
    DOW_map = {
        'MONDAY':    0,
        'TUESDAY':   1,
        'WEDNESDAY': 2,
        'THURSDAY':  3,
        'FRIDAY':    4,
        'SATURDAY':  5,
        'SUNDAY':    6,
    }
    DOW_map1 = {k: np.cos(2*np.pi*v/7.0) for k, v in DOW_map.items()}
    DOW_map2 = {k: np.sin(2*np.pi*v/7.0) for k, v in DOW_map.items()}

    # Show encoding of days of week -> circle
    days = ['MONDAY', 'TUESDAY', 'WEDNESDAY', 'THURSDAY', 'FRIDAY', 'SATURDAY', 'SUNDAY']
    tt = np.linspace(0, 2*np.pi, 200)
    xx = np.cos(tt)
    yy = np.sin(tt)
    plt.plot(xx, yy)
    plt.gca().axis('equal')
    plt.xlabel('Encoded Dimension 1')
    plt.ylabel('Encoded Dimension 2')
    plt.title('2D Projection of days of the week')
    for day in days:
        plt.text(DOW_map1[day], DOW_map2[day], day, ha='center')
    plt.show()
    
    # WEEKDAY_APPR_PROCESS_START to polar coords
    app[col + '_1'] = app[col].map(DOW_map1)
    app[col + '_2'] = app[col].map(DOW_map2)
    app.drop(columns=col, inplace=True)
    return app
application_train_A_reduit = polar_projection(application_train_A_reduit, 'WEEKDAY_APPR_PROCESS_START')
application_test_A_reduit = polar_projection(application_test_A_reduit, 'WEEKDAY_APPR_PROCESS_START')
```

### Boruta

La méthode Boruta est une technique de sélection de variables pour l'apprentissage automatique. Elle s'appuie sur une analyse des relations entre les variables pour déterminer leur importance en vue de la modélisation. Cette méthode utilise des arbres de décision aléatoires pour comparer l'importance des variables avec des versions aléatoires des mêmes variables. Les variables qui sont plus importantes que leurs versions aléatoires sont conservées, tandis que les autres sont supprimées.

Plus précisément, la méthode Boruta suit ces étapes :

1) Elle ajoute des versions aléatoires des variables aux données d'entrée.
2) Elle entraîne un classificateur sur les données d'entrée et calcule l'importance de chaque variable.
3) Elle compare l'importance de chaque variable à celle de ses versions aléatoires, en utilisant un test de significativité.
4) Les variables qui sont plus importantes que leurs versions aléatoires sont considérées comme importantes et sont conservées.
5) Les variables qui ne sont pas plus importantes que leurs versions aléatoires sont considérées comme non importantes et sont supprimées.
6) Les étapes 1 à 5 sont répétées un certain nombre de fois pour améliorer la fiabilité des résultats.
La méthode Boruta peut aider à améliorer les performances des modèles d'apprentissage automatique en réduisant le nombre de variables à prendre en compte, en éliminant les variables qui n'ont pas d'impact significatif sur les résultats. Elle peut être implémentée à l'aide de la bibliothèque Python Boruta-Py.
